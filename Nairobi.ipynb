```python
import pandas as pd

# Replace with your actual filename
df = pd.read_csv('jeans_data.csv')
df.columns = [col.capitalize() for col in df.columns]

# 2. Capitalize first letter of each color
df['Product_name'] = df['Product_name'].str.title()
df['Color'] = df['Color'].str.strip().str.capitalize()
# Preview the first few rows
#print(df.head())
df['Gender'] = df['Gender'].replace({'F': 'Female', 'M': 'Male'})
df = df.rename(columns={'Product_name': 'Product Name', 'Euro_price' : 'Euro Price'})
print(df.columns)
df
```

```python
df.to_csv('jeans_data_cleaned.csv', index=False)
print("✅ CSV with serial numbers saved as 'jeans_data_cleaned.csv'")
```
